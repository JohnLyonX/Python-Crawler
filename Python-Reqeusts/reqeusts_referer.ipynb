```python
import requests
import json
```

```python
# referer 是源地址也就是从https://movie.douban.com/explore这个地址跳转过来的
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
        AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36',
    'referer': 'https://movie.douban.com/explore'
}
res = requests.get('https://m.douban.com/rexxar/api/v2/movie/recommend?refresh=0&start=0&count=20&selected_categories=%7B%7D&uncollect=false&tags=', headers=header)
res.text
```

```python
# result = json.loads(res.text) 不推荐
result = res.json() # 推荐
result
```

```python
for i in res.json()["items"]:
    if i.get('title') is not None:
        print(i.get("title")) # 使用get方法获取字典
        with open("movie_name",'a') as f: # a 表示追加,不会覆盖原来的内容
            f.write(i.get("title")+"\n")
```
